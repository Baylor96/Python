```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = 50
xmax = 1
xmin = -1
x = np.linspace(xmin,xmax,n)
```

```python
def Chebyshev_polynomials(n,x):
    if n == 0:
        k=1
    elif n == 1:
        k=x;
    else :
        k=2*x*Chebyshev_polynomials(n-1,x)-Chebyshev_polynomials(n-2,x);
    return k;
```

```python
c1 = Chebyshev_polynomials(1,x)
c2 = Chebyshev_polynomials(2,x)
c3 = Chebyshev_polynomials(3,x)
c4 = Chebyshev_polynomials(4,x)
c5 = Chebyshev_polynomials(5,x)
c6 = Chebyshev_polynomials(6,x)
```

```python
plt.figure(figsize=(5,5))
plt.title('N>0')
plt.plot(x,c1,label='c1',marker='.')
plt.plot(x,c2,label='c2',marker='.')
plt.plot(x,c3,label='c3',marker='.')
plt.plot(x,c4,label='c4',marker='.')
plt.plot(x,c5,label='c5',marker='.')
plt.plot(x,c6,label='c6',marker='.')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.legend(loc='lower left')
plt.show()
```
